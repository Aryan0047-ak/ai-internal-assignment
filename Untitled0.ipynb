```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
file_path = '/content/Top 10 Crime Cities in US_data.csv'
data = pd.read_csv(file_path)

# Clean the '% of Total Violent Crime' column
data['% of Total Violent Crime'] = data['% of Total Violent Crime'].str.rstrip('%').astype(float)

# Plotting the bar graph with dual y-axes
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot '% of Total Violent Crime' on the primary y-axis
ax1.bar(data['City'], data['% of Total Violent Crime'], color='blue', alpha=0.6, label='% of Total Violent Crime')
ax1.set_ylabel('% of Total Violent Crime', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Create a secondary y-axis for 'Violent Crime'
ax2 = ax1.twinx()
ax2.plot(data['City'], data['Violent Crime'], color='red', marker='o', label='Violent Crime')
ax2.set_ylabel('Violent Crime Rate', color='red')
ax2.tick_params(axis='y', labelcolor='red')

# Adding labels, title, and legend
plt.title('Crime Analysis: Cities vs Crime Statistics')
ax1.set_xlabel('City')
fig.tight_layout()

# Show the plot
plt.show()
```
